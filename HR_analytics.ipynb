```python
# –§—É–Ω–∫–∏—Ü—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –∫—Ä—É–≥–æ–≤—ã—Ö –¥–∏–∞–≥—Ä–∞–º–º
def pie_build(df, column, support_column, title, i, j):
    df_column = df.pivot_table(index = column, values = support_column, aggfunc = ['count'])
    df_column.columns = ['quantity']
    df_column['quantity'] = df_column['quantity'] / df[column].count()
    axes[i, j].set_title(title);
    axes[i, j].pie(df_column['quantity'], autopct='%1.2f%%');
    axes[i, j].legend(
    bbox_to_anchor = (1.0, 1.0),
    loc = 'upper left', labels = df_column.index);
```

```python
# –§—É–Ω–∫–∏—Ü—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –æ–¥–Ω–æ–π –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã
def pie_build_one(df, column, support_column, title):    
    df_column = df.pivot_table(index = column, values = support_column, aggfunc = ['count'])
    df_column.columns = ['quantity']
    df_column['quantity'] = df_column['quantity'] / df[column].count()
    ax.set_title(title);
    ax.pie(df_column['quantity'], autopct='%1.2f%%');
    ax.legend(
    bbox_to_anchor = (1.0, 1.0),
    loc = 'upper left', labels = df_column.index);
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º
def hist_build(df, column, hist_color, xlabel, ylabel, title, i, j, quant_of_bins, more_than_one_row):
    if more_than_one_row:
        axes[i, j].set_title(title) 
        axes[i, j].hist(df[column], bins = quant_of_bins, color = hist_color)
        axes[i, j].set_xlabel(xlabel)
        axes[i, j].set_ylabel(ylabel)
    else:
        axes[j].set_title(title) 
        axes[j].hist(df[column], bins = quant_of_bins, color = hist_color)
        axes[j].set_xlabel(xlabel)
        axes[j].set_ylabel(ylabel)     
```

```python
# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞—Ç–∞—Å–µ—Ç–∞
X_train = train_job_satisf_rt_df.drop(['job_satisfaction_rate', 'id'], axis = 1)
y_train = train_job_satisf_rt_df['job_satisfaction_rate']
```

```python
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer',
             SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
        ),
        (
            'ohe',
             OneHotEncoder(drop = 'first', handle_unknown = 'ignore', sparse_output = False)
        )
    ]
)
```

```python
# SimpleImputer + ORD + SimpleImputer
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord',
            SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories = [
                    ['junior', 'middle', 'senior'],
                    ['low', 'medium', 'high']
                ],
                handle_unknown = 'use_encoded_value', unknown_value = np.nan
            )
                    
        ),
        (
            'simpleImputer_after_ord',
            SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
        )                
    ]       
)
```

```python
# –ü–∞–π–ø–ª–∞–π–Ω –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –∏ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ],
    remainder = 'passthrough'
)
```

<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> 


–û—Ç–ª–∏—á–Ω–æ, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–æ –±–æ–ª–µ–µ –æ–¥–Ω–æ–≥–æ —ç–Ω–∫–æ–¥–µ—Ä–∞, –≤–∑—è—Ç–æ–≥–æ –∏–∑ –≥–æ—Ç–æ–≤–æ–π —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ sklearn)

```python
# –°–ª–æ–≤–∞—Ä—å –¥–ª—è RandomimzedSearchCV
param_distributions = [
            {
                # –û–ø–∏—Å–∞–Ω–∏–µ DecisionTreeClassifier
                'models' : [DecisionTreeRegressor(random_state = RANDOM_STATE)],
                'models__max_depth' : range(5, 13),
                'models__min_samples_split' : range(5, 13),
                'preprocessor__num' : [StandardScaler(), MinMaxScaler(), RobustScaler()]
            },
            {
                # –û–ø–∏—Å–∞–Ω–∏–µ LinearRegression
                'models' : [LinearRegression()
                           ],
                'preprocessor__num' : [StandardScaler(), MinMaxScaler(), RobustScaler()]
            },
]
```

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeRegressor(random_state = RANDOM_STATE)) 
])
```

# –®–∞–≥ 5. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

- –ú–µ—Ç—Ä–∏–∫–∞ SMAPE

```python
def custom_metric(y_true, y_pred):
    error = np.abs(y_true - y_pred)
    scale = (np.abs(y_true) + np.abs(y_pred)) / 2
    return np.mean(error / scale) * 100
```

```python
# –°–æ–∑–¥–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ SMAPE
SMAPE = make_scorer(custom_metric, greater_is_better = False)
```

<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> 
    
–í–µ—Ä–Ω–æ, —á—Ç–æ –∑–∞–¥–∞–µ—à—å –ø–∞—Ä–∞–º–µ—Ç—Ä `greater_is_better=False`. –î–µ–ª–æ –≤ —Ç–æ–º, —á—Ç–æ —Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫–∏ –ø–∞–∫–µ—Ç–∞ `sklearn` –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ —Ñ—É–Ω–∫—Ü–∏–π `cross_val_score` –∏ `GridSearchCV` —É—Ç–≤–µ—Ä–¥–∏–ª–∏ —Å–æ–≥–ª–∞—à–µ–Ω–∏–µ: –≤–æ–∑–≤—Ä–∞—â–∞–µ–º—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤—ã—á–∏—Å–ª–µ–Ω–∏–π –¥–∞–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π –¥–æ–ª–∂–Ω—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –∑–∞–¥–∞—á–µ –º–∞–∫—Å–∏–º–∏–∑–∞—Ü–∏–∏ (—á–µ–º –±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ –ø–æ–ª—É—á–∞–µ–º, —Ç–µ–º –ª—É—á—à–µ). –ü—Ä–∏ —ç—Ç–æ–º –æ–±—Ä–∞—Ç–Ω–∞—è –∑–∞–¥–∞—á–∞ –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Å—É—â–µ—Å—Ç–≤–ª–µ–Ω–∞ —á–µ—Ä–µ–∑ –º–∞–∫—Å–∏–º–∏–∑–∞—Ü–∏—é –º–µ—Ç—Ä–∏–∫–∏, –¥–æ–º–Ω–æ–∂–µ–Ω–Ω–æ–π –Ω–∞ -1. –û–± —ç—Ç–æ–º –º–æ–∂–Ω–æ —á—É—Ç—å –ø–æ–¥—Ä–æ–±–Ω–µ–µ –ø–æ—á–∏—Ç–∞—Ç—å –≤ –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏ –ø–∞–∫–µ—Ç–∞, –ø—Ä–∞–≤–¥–∞ –æ–Ω–∞ –Ω–∞ –∞–Ω–≥–ª–∏–π—Å–∫–æ–º. –û—Å—Ç–∞–≤–ª—é <a href = 'https://scikit-learn.org/stable/modules/model_evaluation.html'>—Å—Å—ã–ª–∫—É</a>. 

- –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

```python
# –û–±—É—á–∏–º –∏ –Ω–∞–π–¥—ë–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–∞ –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞ RandomizedSearchCV
randomized_search = RandomizedSearchCV(
                        pipe_final,
                        param_distributions = param_distributions,
                        cv = 5,
                        scoring = SMAPE,
                        random_state = RANDOM_STATE,
                    )
randomized_search.fit(X_train, y_train)
print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search.best_estimator_)
```

## 3.2 –ü–æ—Ä—Ç—Ä–µ—Ç "—É–≤–æ–ª–∏–≤—à–µ–≥–æ—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞"

```python
# –§—É–Ω–∫–∏—Ü—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –∫—Ä—É–≥–æ–≤—ã—Ö –¥–∏–∞–≥—Ä–∞–º–º
def pie_build_short(df_column, column, title, i, j):
    df_column[column] = df_column[column].apply(lambda x: x / df_column[column].count())
    axes[i, j].set_title(title);
    axes[i, j].pie(df_column[column], autopct='%1.2f%%');
    axes[i, j].legend(
    bbox_to_anchor = (1.0, 1.0),
    loc = 'upper left', labels = df_column.index);
```

```python
def scatterplot_build(df, quant_attribute, target, cat_attribute):
    plt.figure(figsize = (10, 10));
    sns.scatterplot(data = df, x = quant_attribute, y = target, style = cat_attribute, hue = cat_attribute)
    plt.xlabel(quant_attribute);
    plt.ylabel(target);
    ax.legend(
    bbox_to_anchor = (1.0, 1.0),
    loc = 'upper left', labels = df[cat_attribute])
```

<b>–û–±—É—á–∏–ª `LabelEncoder` –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –º–µ—Ç–æ–¥–æ–º `fit_transform`, –∞ –∑–∞—Ç–µ–º —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É –º–µ—Ç–æ–¥–æ–º `transform`. –ù–µ –∑–∞–º–µ—Ç–∏–ª, —á—Ç–æ –≤ –ø—Ä–µ–¥—ã–¥—É—â–∏–π —Ä–∞–∑ –∫ –æ–±–µ–∏–º –≤—ã–±–æ—Ä–∫–∞–º –ø—Ä–∏–º–µ–Ω–∏–ª `fit_transform`</b>

```python
# –ü–∞–π–ø–ª–∞–π–Ω –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –∏ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö
data_preprocessor2 = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns_2),
        ('ord', ord_pipe, ord_columns_2),
        ('num', MinMaxScaler(), num_columns_2)
    ],
    remainder = 'passthrough'
)
```

- –°—Ç—Ä—É–∫—Ç—É—Ä–∞ data_preprocessor —Ç–∞ –∂–µ, –∑–∞–º–µ–Ω–∏–º —Ç–æ–ª—å–∫–æ –∫–æ–ª–æ–Ω–∫–∏ –¥–ª—è –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è –∏ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è

- –î–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ç—Ä–µ–º—è –º–æ–¥–µ–ª—è–º–∏ : LogisticRegression, DecisionTreeClassfier, KNeighborsClassifier

```python
# –°–ª–æ–≤–∞—Ä—å –¥–ª—è RandomimzedSearchCV
param_distributions2 = [
            {
                 # –û–ø–∏—Å–∞–Ω–∏–µ KNeighborsClassifier
                'models' : [KNeighborsClassifier()],
                'models__n_neighbors' : range(4, 8),
                'preprocessor__num' : [StandardScaler(), MinMaxScaler(), RobustScaler()]
            },
            {
                # –û–ø–∏—Å–∞–Ω–∏–µ DecisionTreeClassifier
                'models' : [DecisionTreeClassifier(random_state = RANDOM_STATE)],
                'models__max_depth' : range(5, 12),
                'models__min_samples_split' : range(5, 12),
                'preprocessor__num' : [StandardScaler(), MinMaxScaler(), RobustScaler()]
            },
            {
                # –û–ø–∏—Å–∞–Ω–∏–µ LogisticRegression
                'models' : [LogisticRegression(
                                                random_state = RANDOM_STATE,
                                                solver = 'liblinear',
                                                penalty = 'l1'
                                              )
                           ],
                'models__C' : [0.1, 1, 10],
                'preprocessor__num' : [StandardScaler(), MinMaxScaler(), RobustScaler()]
            }
]
```

```python
pipe_final2 = Pipeline([
    ('preprocessor', data_preprocessor2),
    ('models', DecisionTreeClassifier(random_state = RANDOM_STATE)) 
])
```

# –®–∞–≥ 6. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

```python
# –û–±—É—á–∏–º –∏ –Ω–∞–π–¥—ë–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–∞ –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞ RandomizedSearchCV
randomized_search2 = RandomizedSearchCV(
                        pipe_final2,
                        param_distributions = param_distributions2,
                        cv = 5,
                        scoring = 'roc_auc',
                        random_state = RANDOM_STATE,
                        n_jobs = -1
                    )
randomized_search2.fit(X_train, y_train)
rs2 = randomized_search2.fit(X_train, y_train)
print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search2.best_estimator_)
```

```python
print('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', round((randomized_search2.best_score_), 3))
y_pred = randomized_search2.predict_proba(X_test)[:, 1]
print('–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö:', round(roc_auc_score(y_test, y_pred), 3))
best_model2 = randomized_search2.best_estimator_
```
